```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
metrics = pd.read_csv('Metrics.csv')

metrics_sp500 = metrics[metrics['Stock'] == 'S&P 500']
metrics_tesla = metrics[metrics['Stock'] == 'Tesla']

print(metrics_sp500)
print(metrics_tesla)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(18,6))
fig.suptitle('S&P 500 Metrics Comparison')
colors = ['skyblue', 'salmon', 'mediumaquamarine']
ax1.bar(metrics_sp500['Model'],metrics_sp500['RMSE'], color = colors)
ax1.set_title('RMSE')
ax1.set_ylabel('Score')


ax2.bar(metrics_sp500['Model'], metrics_sp500['MAE'], color = colors)
ax2.set_title('MAE')
ax2.set_ylabel('Score')


ax3.bar(metrics_sp500['Model'], metrics_sp500['MAPE'], color = colors)
ax3.set_title('MAPE')
ax3.set_ylabel('Score in %')


```

```python

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(18,6))
colors = ['skyblue', 'salmon', 'mediumaquamarine']
fig.suptitle('Tesla Metrics Comparison')
ax1.bar(metrics_tesla['Model'],metrics_tesla['RMSE'], color = colors)
ax1.set_title('RMSE')
ax1.set_ylabel('Score')


ax2.bar(metrics_tesla['Model'], metrics_tesla['MAE'], color = colors)
ax2.set_title('MAE')
ax2.set_ylabel('Score')


ax3.bar(metrics_tesla['Model'], metrics_tesla['MAPE'], color = colors)
ax3.set_title('MAPE')
ax3.set_ylabel('Score in %')
```
